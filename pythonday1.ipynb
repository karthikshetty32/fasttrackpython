```python
# Suppose the cover price of a python book  is  577.95, but bookstores get a 40% discount.
# Shipping costs  150 for the first copy and 50 Rupees for each additional copy. What is the total wholesale cost for 60 copies?
book_price = 577.95
discount = 40
ship_of_first_book = 150
ship_additional_book = 50
number_of_copies = 60

cost_of_book = book_price * (1 - (discount/ 100))

cost_of_firstbook = cost_of_book + ship_of_first_book

cost_of_additional = cost_of_book + (number_of_copies - 1) * ship_additional_book

total_wholesale_cost = cost_of_firstbook + cost_of_additional


print(cost_of_book)
print(cost_of_firstbook)
print(cost_of_additional)
print(total_wholesale_cost)
```
